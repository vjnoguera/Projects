```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
import plotly.graph_objects as go
from wordcloud import WordCloud
```

```python
df1 = pd.read_csv(r"C:\Users\Victor\Desktop\EDA\CSV\scimagojr country rank 1996-2023.csv", encoding="latin-1")
df1
```

```python
df1.loc[df1.loc[:,"Documents"].sort_values(ascending=False).index][0:10]
```

```python


# Dividir los valores por 1,000,000 para mostrar en millones
df1["Documents_millions"] = df1["Documents"] / 1000000

# Crear el grÃ¡fico de barras
plt.figure(figsize=(15,8))  # TamaÃ±o ajustado

# Colores progresivos de rojo a azul
colors = sns.color_palette("RdYlBu",18)
# colors = sns.color_palette("flare", as_cmap=True)

barplot = sns.barplot(x=df1["Country"][0:10],
                      y=df1["Documents_millions"][0:10],
                      palette=colors)

# AÃ±adir etiquetas en la parte superior de cada columna
for index, value in enumerate(df1["Documents"][0:10]):
    barplot.text(index, value/1000000, str(value), ha='center', fontsize=12, va='bottom')  # Dividir por 1,000,000 para mostrar en millones

# Ajustar el tamaÃ±o y la inclinaciÃ³n del texto del eje x y y
plt.xticks(rotation=45, ha='right', fontsize=16)
plt.yticks(fontsize=16)

# Etiquetas del eje x y y
plt.xlabel("Country", fontsize=16)
plt.ylabel("Papers (Millions)", fontsize=16)  # Cambiar la etiqueta del eje Y

# TÃ­tulo
plt.title("Top 10 total publications per Country 1996-2023", fontsize=14)

plt.tight_layout()  # Ajuste automÃ¡tico de diseÃ±o
plt.show()
```

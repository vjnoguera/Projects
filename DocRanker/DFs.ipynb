```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
""" CVS descargado de PUBMED y limpio"""
df1 = pd.read_csv("df1_PM2XL_1Klimpio")
# Función para contar palabras
def contar_palabras(texto):
    if pd.isna(texto):  # Si el texto es NaN, retorna 0
        return 0
    return texto.count("asthma")

# Aplicar la función a la columna del DataFrame
df1["word_asthma"] = df1['Abstract'].apply(contar_palabras)
df1.head()
```

```python
""" Dataframe referencia con valores de indicador"""
df2 = pd.read_csv(r"C:\Users\Victor\Desktop\Proyecto ML\Analog\scimagojr 2023_1.csv", sep = "\t")
df2[['Issn1', 'Issn2']] = df2["Issn"].str.split(', ', expand=True)
df2.head()
```

```python
""" CSV descrgado de PUBMED, filtrado y reducido a aquellas filas con abstract"""
df4 = df1[df1["word_asthma"]!=0]
df4.head()
```

```python
""" Dataframe con los resutlados de DF1 filtrando DF2 para poder obtener los rankings de SJR y H index en base a los resultados de DF1"""
# Convertir New_ISSN a un conjunto para búsqueda rápida
new_issn_set = set(df1['New_ISSN'])

# Filtrar df2 donde cualquiera de las columnas Issn1 o Issn2 está en New_ISSN
df5 = df2[df2[['Issn1', 'Issn2']].apply(lambda x: any(issn in new_issn_set for issn in x), axis=1)]
df5.head()
```

```python
""" obtenemos un nuevo dataframe con los valores de df1 dentro de df2, con lo que podemos filtrar por SJR o H index"""
""" Filtramos por SJR"""
list_df5_10_1 = df5["Issn1"].head(10)
list_df5_10_1
```

```python
list_df5_10_2 = df5["Issn2"].head(10)
list_df5_10_2
```

```python
list_df5_10 = pd.concat([list_df5_10_1, list_df5_10_2], ignore_index=True)
list_df5_10
```

```python
result3 = df1[df1['New_ISSN'].isin(list_df5_10)]
result3.head()
```

```python
df6 = df5.loc[df5.loc[:,"H index"].sort_values(ascending=False).index]
df6.head()
```

```python
""" Filtramos por H index"""
list_df6_10_1 = df6["Issn1"].head(10)
list_df6_10_1
```

```python
""" Filtramos por H index"""
list_df6_10_2 = df6["Issn2"].head(10)
list_df6_10_2
```

```python
list_df6_10 = pd.concat([list_df6_10_1, list_df6_10_2], ignore_index=True)
list_df6_10
```

```python
result4 = df1[df1['New_ISSN'].isin(list_df6_10)]
result4.head()
```

```python
""" combinación de los resultados mejores de SJR y H index en un mismo df"""
df7 = pd.merge(result3, result4, on='PMID', how='inner')
df7
```

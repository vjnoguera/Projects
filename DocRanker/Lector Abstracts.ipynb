```python
def clean_abstracts(abstract):
    import re

    sw = ("a", "also","an", "and", "are", "as", "at", "be", "but", "by", "for", "if", "in", "into", "is", "it", "its", "may", "no",
          "not", "of", "on", "or", "such", "that", "the", "their", "then", "there", "these", "they", "this", "to", "was", "will", "with")

    abstract = re.sub(r"-", " ", abstract)
    abstract = re.sub(r"\d+", "{num}", abstract)
    abstract = re.sub(r"[^\w\s]", "", abstract)
    abstract = " ".join([word for word in abstract.lower().split() if word not in (sw)])
    return abstract.lower()
```

```python
df3["clean_Abstract"] = df3["Abstract"].apply(lambda x: clean_abstracts(x))
```

```python
""" modelo propuesto"""

from sentence_transformers import SentenceTransformer, util
modeli = SentenceTransformer('all-MiniLM-L6-v2')

#Our sentences we like to encode
sentences = df3['clean_Abstract'].values

#Sentences are encoded by calling model.encode()
embeddings = modeli.encode(sentences)
```

```python
embeddings.shape
```

```python
from sentence_transformers import util
cosine_scores = util.cos_sim(embeddings, modeli.encode(df3["clean_Abstract"].iloc[0]))
```

```python

from sklearn.metrics.pairwise import cosine_similarity
# Compute cosine similarity between the first sample and all samples
similarity_matrix = cosine_similarity(embeddings)
similarities_to_first = similarity_matrix[0]

# Find the indices of the top 10 most similar documents
top_10_indices = np.argsort(similarities_to_first)[-11:-1][::-1]  # Exclude the first sample itself

print("Indices of top 10 most similar documents to the first sample:", top_10_indices)
print("Similarity scores of top 10 most similar documents:", similarities_to_first[top_10_indices])
```
